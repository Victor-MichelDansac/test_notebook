```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x,f):
    return( np.sin( f * x * 2*np.pi ) )
```

```python
N = 100
x = np.linspace(0,1,N)
```

```python
plt.plot(x,func(x,2))
plt.show()
```

```python
from ipywidgets import interact
```

```python
def plot_func(f):
    plt.plot(x,func(x,f))
    plt.show()
```

```python
interact( plot_func, f=(0,10,0.5) )
```
